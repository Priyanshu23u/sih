```python
import pandas as pd

# Load the dataset
df = pd.read_csv("news_data1.csv")

# Define the list of crime-related keywords
disaster_keywords = {
    'earthquake': 'Earthquake', 
    'flood': 'Flood', 
    'tsunami': 'Tsunami', 
    'hurricane': 'Hurricane', 
    'tornado': 'Tornado', 
    'wildfire': 'Wildfire',
    'landslide': 'Landslide', 
    'volcano': 'Volcanic Eruption', 
    'drought': 'Drought', 
    'cyclone': 'Cyclone', 
    'storm': 'Storm', 
    'avalanche': 'Avalanche', 
    'blizzard': 'Blizzard',
    'heatwave': 'Heatwave', 
    'hailstorm': 'Hailstorm', 
    'mudslide': 'Mudslide', 
    'famine': 'Famine', 
    'pandemic': 'Pandemic', 
    'outbreak': 'Outbreak',
    'snowstorm': 'Snowstorm', 
    'sandstorm': 'Sandstorm', 
    'disaster': 'Disaster', 
    'natural disaster': 'Natural Disaster', 
    'calamity': 'Calamity', 
    'emergency': 'Emergency',
    'hazard': 'Hazard', 
    'catastrophe': 'Catastrophe', 
    'rescue': 'Rescue Operation',
    'relief': 'Relief Effort',
    'evacuation': 'Evacuation',
    'debris': 'Debris',
    'tsunami warning': 'Tsunami Warning',
    'aftershock': 'Aftershock',
    'lightening':'lightening',
}

# Function to label data based on keywords
def label_data(title):
    for keyword in disaster_keywords:
        if keyword in title.lower():
            return 1  
    return 0  


# Apply the functions to the dataset
df['label'] = df['Title'].apply(label_data)
df.to_csv('ans.csv', index=False)

# Drop duplicates
```

```python
df = df.drop_duplicates()

# Filter for rows with crime-related titles
df_zero_label = df[df['label'] == 1]
df_zero_label= df_zero_label.drop_duplicates()
# Save the new dataset with crime type detection
df.to_csv('ans1.csv', index=False)

# Display the result
print(df_zero_label)
```
